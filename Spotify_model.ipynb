```python
for col in dataset.columns:
    if dataset[col].dtypes == "object":
        dataset = dataset.drop(labels = [col],axis = 1)
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(data=dataset.corr(),annot=True)
plt.show()
```

```python
y = dataset["mode"]
X = dataset.drop(labels = "mode",axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()
X = mms.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
gnb.fit(X_train,y_train)
y_predict_gnb = gnb.predict(X_test)
```

```python
print(classification_report(y_predict_gnb,y_test))
print(accuracy_score(y_predict_gnb,y_test))
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()
dtc.fit(X_train,y_train)
y_predict_dtc = dtc.predict(X_test)
```

```python
print(classification_report(y_predict_dtc,y_test))
print(accuracy_score(y_predict_dtc,y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train,y_train)
y_predict_rfc = rfc.predict(X_test)
```

```python
print(classification_report(y_predict_rfc,y_test))
print(accuracy_score(y_predict_rfc,y_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3,metric="euclidean")
knn.fit(X_train,y_train)
y_predict_knn = knn.predict(X_test)
```

```python
print(classification_report(y_predict_knn,y_test))
print(accuracy_score(y_predict_knn,y_test))
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(64, input_dim=14, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=150, batch_size=10)
```

```python
_, accuracy = model.evaluate(X_test, y_test)
```
